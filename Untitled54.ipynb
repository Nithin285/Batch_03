```python
import numpy as np
x=np.array([[2,4,3],[3,6,5],[5,7,7],[6,8,8],[1,3,2],[7,7,9]])
y=np.array([0,0,1,1,0,1])
```

```python
W=np.zeros(x.shape[1])
lr=1
epochs=6
b=0
```

```python
def step_function(val):
  return 1 if val>=0 else 0
```

```python
for epoch in range(epochs):
    print(f"epoch:{epoch+1}")
    for i in range(len(x)):
        z=np.dot(x[i],W)+b
        yp=step_function(z)
        error=y[i]-yp
        # Update weights and bias only if there's an error
        if error != 0:
            W=W+lr*error*x[i]
            b=b+lr*error
            print(f"  Input: {x[i]}, Actual: {y[i]}, Predicted: {yp}, Error: {error}")
            print(f"  Updated Weight: {W}, Updated Bias: {b}")
        else:
            print(f"  Input: {x[i]}, Actual: {y[i]}, Predicted: {yp}, No update (error = 0)")
    print(f"End of epoch {epoch+1}. Current Weight: {W}, Current Bias: {b}")
print("\nTraining complete.")
print(f"Final Weight: {W}")
print(f"Final Bias: {b}")
```
