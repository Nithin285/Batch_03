```python
import numpy as np
x=np.array([[5,3,1],[1,0,1],[7,5,1],[2,1,1],[6,4,1]])
y=np.array([1,0,1,0,1])
```

```python
W=np.zeros(x.shape[1])
lr=1
epochs=5
b=0
```

```python
def step_function(val):
  return 1 if val>=0 else 0
```

```python
for epoch in range(epochs):
    print(f"epoch:{epoch+1}")
    for i in range(len(x)):
        z=np.dot(x[i],W)+b
        yp=step_function(z)
        error=y[i]-yp
        if error==0:
            W=W+lr*(y[i]-yp)*x[i]
            b=b+lr*(y[i]-yp)
            print(f"input {x[i]} Actual value {y[i]} Predicted value : {yp}")

            print("final Weight",W)
            print("final Bias",b)
```
